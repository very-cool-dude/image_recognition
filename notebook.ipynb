# Задание 3

Была создана вспомогательная функция для поиска количества пикселей, которые следует зашумить. На вход она принимает матрицу образа и процент зашумления.

```python
def find_number_of_pixels_to_make_noisy(image, percent):
  total = 0
  for i in image:
    total += len(i)
  total *= percent/100
  return int(total)
```

Была создана функция, которая получает на вход матрицу образа и процент зашумления. В процессе она обращается к функции find_number_of_pixels_to_make_noisy(), для поиска количества пикселей для изменения. Дальше функция случайным образом выбирает строчку и столбец матрицы и меняет значение на противоположное. Если значение уже было изменено, то выбирается новое случайное значение. Это сделано затем, чтобы более наглядно работало зашумление с разным процентом.

```python
def make_some_noise(image, percent):
  exceptions = []
  noisy_image = np.copy(image)
  noisy_pixels = find_number_of_pixels_to_make_noisy(image, percent)
  for i in range(noisy_pixels):
    j = randrange(len(image))
    k = randrange(len(image[j]))
    while [j, k] in exceptions:
      j = randrange(len(image))
      k = randrange(len(image[j]))
    noisy_image[j][k] *= -1
    exceptions.append([j , k])
  noisy_arrow_image = noisy_image.flatten()
  return noisy_arrow_image
```

Была создана функция, которая получает на вход количество генерируемых изображений с шумами и дальше, используя предыдущие функции, создаёт список векторов с шумами.

```python
def make_more_noise(x_arrows = x_arrows, x_matrix = x_matrix, N = 100):
  noisy_arrows = []
  for i in range(1, N+1):
    random_number = i
    if N != 100:
      random_number = randint(1, 100)
    noisy_image_label = choice(list(x_arrows.keys()))
    noisy_arrows.append({"Вектор с шумами":make_some_noise(x_matrix[noisy_image_label], random_number),"Вектор после предсказания":'', "Изображение без шумов":noisy_image_label, "Предсказанное изображение":'', "Процент зашумления":random_number, "Ошибка":''})
  return noisy_arrows
```

Был создан набор данных с шумами. Он состоит из 100 элементов.

# Задание 4

Была создана функция для нахождения матрицы весовых коэффициентов. Данная функция создаёт копию эталонной матрицы и создаёт её транспонированную версию. Функция возвращает произведение полученных матриц.

```python
def find_value_matrix(image):
  array = np.copy(image)
  array_T = np.transpose(array)
  return np.outer(array, array_T)
```

Каждый эталонный вектор был преобразован в матрицу весовых коэффициентов. В данном случае их три - по количеству образов.

```python
circle_dot = find_value_matrix(circle_arrow)
square_dot = find_value_matrix(square_arrow)
triangle_dot = find_value_matrix(triangle_arrow)
```

Матрицы были просуммированы. Была найдена длина вектора.

```python
W = circle_dot + square_dot + triangle_dot
```

```python
N = len(circle_arrow)
```

Матрица была поделена на длину вектора

```python
W = W / N
```

Была создана функция для обнуления диагональных элементов. Данная функция создаёт копию входной матрицы и заменяет диагональные ячейки на ноль.

```python
def make_diagonal_zero(value_matrix):
  zero_matrix = np.copy(value_matrix)
  for i in range(len(value_matrix)):
      zero_matrix[i][i] = 0
  return zero_matrix
```

Произошло обнуление диагональных элементов суммарной матрицы

# Задание 5

Была создана функция, которая меняет отрицательные числа на -1, а неотрицатльные на 1.

```python
def make_plus_or_minus(array):
  for i in range(len(array)):
    if array[i] < 0:
      array[i] = -1
    elif array[i] >= 0:
      array[i] = 1
```

Была создана функция, которая находит и вносит словарь зашумлённых изображений значение предсказанного образа. Если совпадение, то будет написан незашумлённый образ, а если ошибка, то фраза "false prediction"

```python
def get_prediction_image_origin(x_arrows = x_arrows, noisy_arrows = noisy_arrows):
  for j in noisy_arrows:
    for i in x_arrows:
      if np.array_equal(j['Вектор после предсказания'], x_arrows[i]):
        j['Предсказанное изображение'] = i
        break
      else:
        j['Предсказанное изображение'] = 'false prediction'
```

Была создана функция предсказания, используя предыдущие функции. Функция проходит по всем зашумлённым векторам и умножает их на суммарную матрицу W, затем заменяет значения меньше нуля на -1, а неотрицательные - на 1.

```python
def make_prediction(noisy_arrows = noisy_arrows, W = W, x_arrows = x_arrows):
  for i in noisy_arrows:
    y = i['Вектор с шумами']
    y = np.dot(W, y)
    make_plus_or_minus(y)
    i['Вектор после предсказания'] = y
  get_prediction_image_origin(x_arrows, noisy_arrows)
  return
```

Было произведено предсказание для всех зашумлённых векторов.

```python
make_prediction()
```

## Средние значения для процента угадывания и ошибки

Была проверена разница изначальной и новой модели на 100 итерациях. Были записаны средние значения ошибок для каждой итерации и значения отношения ложно угаданных образов. По данным метрикам были посчитаны средние значения.

```python
errors = []
errors_dop = []

percent = []
percent_dop = []

N = 100

for i in range(N):
  def fill_dop_metrics():
    noisy_arrows_dop = make_more_noise(x_arrows_dop, x_matrix_dop)
    make_prediction(noisy_arrows_dop, W_dop, x_arrows_dop)
    false_predict_array_dop = []
    print_false_predict(noisy_arrows_dop, false_predict_array_dop, False)
    percent_dop.append(len(false_predict_array_dop))
    find_and_add_errors(noisy_arrows_dop, x_arrows_dop)

    temp_errors = []

    for i in false_predict_array_dop:
      temp_errors.append(i['Ошибка'])

    errors_dop.append(sum(temp_errors) / len(temp_errors))

  def fill_vanilla_metrics():
    noisy_arrows = make_more_noise()
    make_prediction(noisy_arrows)
    false_predict_array = []
    print_false_predict(noisy_arrows, false_predict_array, False)
    percent.append(len(false_predict_array))
    find_and_add_errors(noisy_arrows)

    temp_errors = []

    for i in false_predict_array:
      temp_errors.append(i['Ошибка'])

    errors.append(sum(temp_errors) / len(temp_errors))

  fill_vanilla_metrics()
  fill_dop_metrics()

mean_errors = sum(errors) / len(errors)
mean_errors_dop = sum(errors_dop) / len(errors_dop)
mean_percent = sum(percent) / len(percent)
mean_percent_dop = sum(percent_dop) / len(percent_dop)

print('Средняя ошибка для базовой модели', mean_errors)
print('Средняя ошибка для дополненной модели', mean_errors_dop)
print(50 * '-')
print('Средний процент ложных предсказаний для базовой модели', mean_percent)
print('Средний процент ложных предсказаний для дополненной модели', mean_percent_dop)
```

По данным метрикам можно сделать вывод, что новая модель ошибается чаще, чем базовая, но разница между набольшей и наименьшей ошибкой меньше, значит диапазон значений новой модели уже, чем у базовой.
